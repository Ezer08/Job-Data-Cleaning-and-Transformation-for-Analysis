**Drop the 'index' column**

```python
dfc=df.copy()
```

```python
# Drop the 'index' column
dfc= dfc.drop(columns=['index'])

# Verify that the column has been removed
print(dfc.columns)
```

**Extract from Salary Estimate: Minimum, Maximum, and Average Salary**

```python
# Remove unnecessary characters like 'K', '$', and 'Glassdoor est.' in dfc
dfc['Salary Estimate'] = dfc['Salary Estimate'].str.replace('K', '').str.replace('$', '').str.replace('(Glassdoor est.)', '').str.strip()

# Split the salary range into two columns in dfc
dfc[['Salary Min', 'Salary Max']] = dfc['Salary Estimate'].str.split('-', expand=True)

```

```python
dfc.head(5)
```

```python
dfc.dtypes
```

```python
# Convert Salary Min and Salary Max to integers if needed
# Extract numeric part using regex
# Ensure the columns are treated as strings first
dfc['Salary Min'] = dfc['Salary Min'].astype(str).str.extract('(\d+)').astype(int)
dfc['Salary Max'] = dfc['Salary Max'].astype(str).str.extract('(\d+)').astype(int)

# Check the updated DataFrame to ensure everything looks correct
print(dfc[['Salary Min', 'Salary Max']])
```

```python
print(dfc.dtypes)
```

```python
# Rename the 'Salary Min' and 'Salary Max' columns
dfc = dfc.rename(columns={'Salary Min': 'Salary Min estimated', 'Salary Max': 'Salary Max estimated'})

# Calculate the average salary and create a new column called 'average salary estimated'
dfc['average salary estimated'] = (dfc['Salary Min estimated'] + dfc['Salary Max estimated']) / 2

# Check the updated DataFrame
print(dfc[['Salary Min estimated', 'Salary Max estimated', 'average salary estimated']])
```

**Clean the Job Description column**

```python
# Clean the Job Description column by removing newline characters and extra whitespace
dfc['Job Description'] = dfc['Job Description'].str.replace('\n', ' ').str.strip()

# Check the updated Job Description column
print(dfc['Job Description'])
```

**Clean the Company Name**

```python
# Clean the Company Name column by removing the rating
dfc['Company Name'] = dfc['Company Name'].str.split('\n').str[0]

# Check the updated Company Name column
print(dfc['Company Name'])
```

**Extract the state from the Location column**

```python
# Extract the state from the Location column
dfc['State'] = dfc['Location'].str.split(',').str[-1].str.strip()

# Check the updated DataFrame
print(dfc[['Location', 'State']])
```

```python
# Get the unique states from the State column
unique_states = dfc['State'].unique()

# Print the list of unique states
print(unique_states)
```

**Clean column Size**

```python
# Rename the Size column to Number of Employees
dfc.rename(columns={'Size': 'Number of Employees'}, inplace=True)

# Check the updated DataFrame
print(dfc[['Number of Employees']])
```

```python
# Function to convert number of employees to a list of integers
def convert_employees(emp_str):
    # Check for 'Unknown' and handle it
    if 'Unknown' in emp_str:
        return None  # or return [] if you prefer an empty list

    # Remove ' employees' and handle cases with '+' or any other characters
    emp_str = emp_str.replace(' employees', '').replace('+', '').strip()

    # Split the string on ' to ' and convert to integers
    if ' to ' in emp_str:
        return [int(x) for x in emp_str.split(' to ')]
    else:
        # Handle cases with a single number
        return [int(emp_str), int(emp_str)]

# Apply the conversion function to the Number of Employees column
dfc['Number of Employees'] = dfc['Number of Employees'].apply(convert_employees)

# Check the updated Number of Employees column
print(dfc[['Number of Employees']])
```

```python
# Function to convert revenue to a list of strings in the desired format
def convert_revenue(revenue_str):
    # Remove the '(USD)' part and any leading/trailing whitespace
    revenue_str = revenue_str.replace(' (USD)', '').strip()

    # Check for 'Unknown' and handle it
    if 'Unknown' in revenue_str:
        return None  # or return [] if you prefer an empty list

    # Split the string on ' to ' and clean each part
    if ' to ' in revenue_str:
        # Create a list and ensure it maintains the '$' sign and ' billion'
        return [part.strip() + ' billion' for part in revenue_str.split(' to ')]
    else:
        # Handle cases with a single value
        return [revenue_str.strip() + ' billion', revenue_str.strip() + ' billion']

# Apply the conversion function to the Revenue column
dfc['Revenue'] = dfc['Revenue'].apply(convert_revenue)

# Rename the Revenue column to Revenue Range
dfc.rename(columns={'Revenue': 'Revenue Range'}, inplace=True)

# Check the updated DataFrame
print(dfc[['Revenue Range']])
```

```python
# Replace None values in the Revenue Range column with 'Unknown'
dfc['Revenue Range'] = dfc['Revenue Range'].apply(lambda x: x if x is not None else 'Unknown')

# Check the updated Revenue Range column
print(dfc[['Revenue Range']])
```
